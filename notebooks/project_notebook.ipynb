```python
import pandas as pd

# Assuming 'updated_gdp_data.csv' and 'another_file.csv' are in the current directory
# Replace with your actual file paths if needed.
file1_path = '/content/drive/MyDrive/CAV/PROJECT-3/energy consumption.csv'
file2_path = '/content/drive/MyDrive/CAV/PROJECT-3/gdp_data.csv'

try:
    df1 = pd.read_csv(file1_path)
    df2 = pd.read_csv(file2_path)

    # Merge the two dataframes based on a common column.
    # Replace 'common_column' with the actual name of the column you want to merge on.
    merged_df = pd.merge(df1, df2, on='Country Name', how='outer')

    print(merged_df)

except FileNotFoundError:
    print(f"One or both of the files ({file1_path}, {file2_path}) were not found.")
except KeyError as e:
    print(f"Error: The specified merge key '{e}' was not found in one or both dataframes. Please check your column names.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
